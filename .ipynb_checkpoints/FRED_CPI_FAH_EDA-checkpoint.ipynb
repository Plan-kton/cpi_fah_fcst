```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load CSV file
df = pd.read_csv('fetch_fred_data.csv')
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
df = df[df['date'] >= '2004-01-01']
df = df.reset_index().set_index('date').drop(columns = 'index')

df
```

```python
df_cpi = df[["US Grocery Sales", "CPI (Food at Home)", "Restaurant Sales", "Oil Prices", "PPI Farm Products", "PPI Food Manufacture",
            "PPI Grocery", "Retail Wages"]]

df_cpi = df_cpi.rename(columns={
        "US Grocery Sales": 'grocery_sales',  
        "CPI (Food at Home)": 'cpi_fah', 
        "Restaurant Sales": 'restaurant_sales',
        "Oil Prices": 'oil_prices',        
        "PPI Farm Products": 'ppi_farm_products',
        "PPI Food Manufacture": 'ppi_food_mfg',
        "PPI Grocery": 'ppi_grocery',
        "Retail Wages": 'wages_retail'
    }
)

df_cpi['grocery_units'] = df_cpi['grocery_sales']/df_cpi['cpi_fah']
```

```python
# Identify outliers using IQR
cols = df_cpi.columns.tolist()

Q1 = df_cpi[cols].quantile(0.25)
Q3 = df_cpi[cols].quantile(0.75)

IQR = Q3 - Q1

lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR

IQR_df_cpi = pd.DataFrame(
    {'Column': Q1.index,                        # Column names
    'Q1 (25th percentile)': Q1.values,          # Q1 values
    'Q3 (75th percentile)': Q3.values,          # Q3 values
    'IQR': IQR.values,                          # Interquartile range
    'Lower Whisker': lower_whisker.values,      # Lower whisker
    'Upper Whisker': upper_whisker.values})      # Upper whisker

#This provides the percent of values above or below the whiskers
((df_cpi[cols] < lower_whisker) | (df_cpi[cols] > upper_whisker)).sum() / len(df_cpi) * 100
```

### Visualizations ###

```python
# Select the variables to plot (excluding 'date')
var_list = df_cpi.columns.tolist()

# Create a grid of subplots (2 columns × enough rows for all variables)
num_vars = len(var_list)
rows = (num_vars // 2) + (num_vars % 2)  # Ensures enough rows for all variables

fig, axes = plt.subplots(nrows=rows, ncols=2, figsize=(15, 5 * rows))  # Dynamically sized

# Flatten axes array for easier iteration
axes = axes.flatten()

# Loop through each variable and plot
for i, col in enumerate(var_list):
    ax = axes[i]  # Select the subplot
    ax.plot(df_cpi.index, df_cpi[col], label=col, linewidth=2)
    
    # Customize the subplot
    ax.set_title(col, fontsize=12)
    ax.set_xlabel("Date", fontsize=10)
    ax.set_ylabel("Value", fontsize=10)
    ax.grid(True)
    ax.legend()

# Adjust layout for readability
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Select independent variables (excluding index)
var_list = df_cpi.columns.tolist()  # Since 'date' is index, it's not included

# Create a grid of subplots (2 columns × enough rows for all variables)
num_vars = len(var_list)
rows = (num_vars // 2) + (num_vars % 2)  # Ensures enough rows for all variables

fig, axes = plt.subplots(nrows=rows, ncols=2, figsize=(15, 5 * rows))  # Dynamically sized

# Flatten axes array for easier iteration
axes = axes.flatten()

# Loop through each variable and create scatter plots
for i, col in enumerate(var_list):
    ax = axes[i]
    ax.scatter(df_cpi[col], df_cpi['cpi_fah'], color='red', alpha=0.6)
    ax.set_xlabel(col, fontsize=10)
    ax.set_ylabel('Food Inflation', fontsize=10)
    ax.set_title(f'Scatter Plot: {col} vs Food Inflation')

# Hide empty subplots if `var_list` has an odd number of elements
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout and show all plots
plt.tight_layout()
plt.show()
```

### Feature Creation ###

```python
#feature creation

# Create first difference variables (applies to all numeric columns)
df_cpi_diff = df_cpi.diff().add_suffix('_diff')
df_cpi_diff = df_cpi_diff.bfill()

# Create YOY variables (applies to all numeric columns)
df_cpi_yoy = df_cpi.pct_change(periods=12, fill_method=None).add_suffix('_yoy') * 100  # Converts to percentage

# Create dummy1 for dates between 2/29/2020 and 4/30/2020. A boolean consition with a .astype(int) will return a 1 for True and 0 for false
df_cpi['covid1'] = ((df_cpi.index >= '2020-02-01') & (df_cpi.index <= '2020-04-01')).astype(int)

# Create dummy2 for dates between 5/31/2020 and 10/31/2020
df_cpi['covid2'] = ((df_cpi.index >= '2020-05-01') & (df_cpi.index <= '2020-10-01')).astype(int)
```

```python
df_cpi = pd.concat([df_cpi, df_cpi_diff, df_cpi_yoy], axis=1)
```

### Lag Variable Analysis ###

```python
# 1) create lagged variables
lag_vars = ['cpi_fah_yoy', 'oil_prices_yoy', 'ppi_farm_products_yoy', 'ppi_food_mfg_yoy', 'ppi_grocery_yoy', 'wages_retail_yoy']

lags = range(1, 13)

lagged_cols = {}

for lag_var in lag_vars:
    for lag in lags:
        col_name = f'{lag_var}_lag{lag}'
        lagged_cols[col_name] = df_cpi[lag_var].shift(lag)

# Add all lagged columns at once using concat
df_cpi = pd.concat([df_cpi, pd.DataFrame(lagged_cols, index=df_cpi.index)], axis=1)
```

```python
lags = range(1, 13)

# 1) define vars
ppi_vars = ['oil_prices_yoy', 'ppi_farm_products_yoy', 'ppi_food_mfg_yoy', 'ppi_grocery_yoy', 'wages_retail_yoy']
target_var = 'cpi_fah_yoy'

# Dictionary to store results
correlation_results = {}

for ppi_var in ppi_vars:
    lag_cols = []
    for lag in lags:
        lag_col = f'{ppi_var}_lag{lag}'
        lag_cols.append(lag_col)

    # Compute correlation matrix with lagged variables
    corrs = df_cpi[[target_var] + lag_cols].corr()
    corr_values = corrs.loc[target_var, lag_cols]
    correlation_results[ppi_var] = corr_values.values

# Convert to DataFrame
corr_df_cpi_ppi = pd.DataFrame(correlation_results, index=[f'lag{lag}' for lag in lags])

# 2) Plot the results
plt.figure(figsize=(10, 6))

for i, column in enumerate(corr_df_cpi_ppi.columns):
    plt.plot(lags, corr_df_cpi_ppi.iloc[:, i], marker='o', linewidth=2,
             label=column.replace('_yoy', '').replace('_', ' ').title())

plt.title('Correlation of CPI Food at Home with Lagged PPI Variables')
plt.xlabel('Lag (months)')
plt.ylabel('Correlation Coefficient')
plt.xticks(lags)
plt.ylim(-0.25, 1)
plt.legend(title="Independent Variables", loc='lower left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
